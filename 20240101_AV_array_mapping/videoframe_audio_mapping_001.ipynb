```python
import easygui
import pandas as pd
import cv2
import numpy as np
import librosa

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)
```

```python
image_array = easygui.fileopenbox(msg='Select the image files', title='Select the image files', default='/media/user7/Double_Dragon/Videography/My_Archive/', filetypes=['*.jpg', '*.png', '*.jpeg', '*.JPG', '*.JPEG'], multiple=True)
```

## Image arrays to DataFrame

```python
df = pd.DataFrame(columns=["cv2_image", "filepath"])

for filepath in image_array:
    cv2_image = cv2.imread(filepath)
    cv2_image = cv2.cvtColor(cv2_image, cv2.COLOR_BGR2RGB)

    df = df.append(
        {"cv2_image": cv2_image, "filepath": filepath},
        ignore_index=True)

df
```
