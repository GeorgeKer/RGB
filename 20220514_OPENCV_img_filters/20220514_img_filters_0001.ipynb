```python
import cv2
import scipy
import numpy as np
import matplotlib.pyplot as plt
import  ipywidgets
from PIL import Image
import math

INPUTFILE_PATH = "/home/user7/Pictures/313198810_n.jpg"
# OUTPUT_PATH = "/OUTPUT/" + INPUTFILE_PATH.split('/')[-2].split(".")[0] + "_001.jpg"
INIT_IMG = cv2.imread(INPUTFILE_PATH)
INIT_IMG = cv2.cvtColor(INIT_IMG, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(13, 13))

plt.imshow(INIT_IMG)
```

```python
plt.figure(figsize=(13, 13))

FVAL = 80

mask = INIT_IMG < FVAL
out_img = INIT_IMG.copy()
out_img[mask] = 255 - out_img[mask] + math.floor(1.5*FVAL)

plt.imshow(out_img)
# print(filter)
```
